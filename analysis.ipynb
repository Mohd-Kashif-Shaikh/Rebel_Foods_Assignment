```python
import pandas as pd
```

```python
df = pd.read_csv("data//Kittchen.csv")
```

```python
df.head()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df["MONTH"] = pd.to_datetime(df["MONTH"],format="%b-%y")
df.head()
```

```python
df = df.sort_values("MONTH")
df.head()
```

```python
df["GM_PCT"]=(df["GROSS MARGIN"]/df["NET REVENUE"])*100
df.head()
```

```python
df["EBITDA_PCT"]=(df["KITCHEN EBITDA"]/df["NET REVENUE"])*100
df.head()
```

```python
df["AOV"]=(df["NET REVENUE"]/df["ORDER COUNT"])
```

```python
df["VARIANCE_PCT"]=(df["VARIANCE"]/df["IDEAL FOOD COST"])*100
```

```python
df["VARIANCE_PCT"].describe()
```

```python
df["VARIANCE_PCT"].describe()
```

```python
variance_bins = [0, 2, 3, 5, 100]
variance_labels = ["Var <2%","Var 2%-3%","Var 3%-5%","Var >5%"]
df["VARIANCE_CATEGORY"] = pd.cut(df["VARIANCE_PCT"],bins=variance_bins,labels=variance_labels,include_lowest=True)
```

```python
df.head()
```

```python
df["VARIANCE_CATEGORY"].value_counts()
df.head()
```

```python
df["DISCOUNT_PCT"]=(df["DISCOUNT"]/df["CART SALES"])*100
df.head()
```

```python
##Discount vs EBITDA
import plotly.express as px
fig = px.scatter(cd, x="DISCOUNT",y="EBITDA_PCT",color="CITY",hover_data=["STORE"],title="Discount vs EBITDA %"
)
fig.show()
```

```python
fig=px.scatter(cd,x="DISCOUNT_PCT",y="EBITDA_PCT",color="CITY",hover_data=["STORE"],trendline="ols",
title="Discount % vs EBITDA %"
)
fig.show()
```

```python
#How is business revenue changing month-over-month?
monthly_revenue = (
    cd.groupby("MONTH")["NET REVENUE"]
    .sum()
    .reset_index()
)
```

```python
fig = px.line(
    monthly_revenue,
    x="MONTH",
    y="NET REVENUE",
    markers=True,
    title="Monthly Revenue Trend"
)
fig.show()
```

```python
#Is profitability improving over time?
monthly_ebitda = (
    df.groupby("MONTH")["KITCHEN EBITDA"]
    .sum()
    .reset_index()
)
```

```python
fig = px.line(
    monthly_ebitda,
    x="MONTH",
    y="KITCHEN EBITDA",
    markers=True,
    title="Monthly EBITDA Trend"
)

fig.show()
```

```python
#Which cities are most profitable?
city_ebitda = (
    df.groupby("CITY")["KITCHEN EBITDA"]
    .sum()
    .sort_values(ascending=False)
    .reset_index()
)
```

```python
fig = px.bar(
    city_ebitda,
    x="CITY",
    y="KITCHEN EBITDA",
    title="City-wise EBITDA"
)

fig.show()
```

```python
#Which stores contribute most to profitability?
top_stores = (
    df.groupby("STORE")["KITCHEN EBITDA"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
    .reset_index()
)
```

```python
fig = px.bar(
    top_stores,
    x="KITCHEN EBITDA",
    y="STORE",
    orientation="h",
    title="Top 10 Stores by EBITDA"
)

fig.show()
```

```python
#Does food wastage reduce profitability?
fig = px.scatter(
    df,
    x="VARIANCE_PCT",
    y="EBITDA_PCT",
    color="CITY",
    hover_data=["STORE"],
    title="Variance % vs EBITDA %"
)

fig.show()
```
